# 15.GÖZ İZLEME (EYE MOTİON TRACKİNG)

```python
import cv2
import numpy as np


cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()
    if ret is False:
        break
    
    #frame = cv2.resize(frame,(640,480))
    roi = frame[80:210,230:450]
    
    rows,cols,_ = roi.shape
    
    print(roi.shape)
    gray = cv2.cvtColor(roi,cv2.COLOR_BGR2GRAY)
    ret,threshold = cv2.threshold(gray,38,255,cv2.THRESH_BINARY_INV)
    
    contours,ret = cv2.findContours(threshold,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    contours = sorted(contours, key = lambda x: cv2.contourArea(x),reverse = True)  # sorted fonksiyonu içine girilen değerleri belirli bir sıraya  göre dizmemizi sağlar neye göre sıralayacağını 2. parametre olarak girdik, 3. girdiğimiz parametre ise tersten sıralamasını sağlar yani küçükten büyüğe değil de büyükten küçüğe sıralamasını sağlar
    
    for cnt in contours: 
        (x,y,w,h) = cv2.boundingRect(cnt) # cv2.boundingRect() fonksiyonu bir dikdörtgen çizebilmek için ihtiyacımız olan x,y, ve w,h değerlerini verir
        cv2.rectangle(roi,(x,y),(x+w,y+h),(255,0,0),thickness=2)
        cv2.line(roi,(x+int(w/2),0),(x+int(w/2),rows),(0,255,0),2)
        cv2.line(roi,(0,y+int(h/2)),(cols,y+int(h/2)),(0,255,0),2)
        break
        
    cv2.imshow("frame",frame)
    cv2.imshow("roi",roi)
    cv2.imshow("threshold",threshold)
    
    if  cv2.waitKey(80) == ord('q'):
        break
  
cap.release()
cv2.destroyAllWindows()
```
